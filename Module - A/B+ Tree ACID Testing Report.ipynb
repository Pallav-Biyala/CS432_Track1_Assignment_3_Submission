 <h2 style=" font-family:'Times New Roman';"><b>Atomicity Test:</b></h2>

<h3 style=" font-family:'Times New Roman';">Atomicity means that a transaction must execute completely or not at all. In simple words: A transaction performs multiple operations. If one operations fails, then all previous operations must be undone. Hence, database should never be left partially updated.</h3>

<h3 style=" font-family:'Times New Roman';">For example: If a transaction does:
<ol>
    <li>Update student table</li>
    <li>Update course table</li>
    <li>Insert in enrollment table</li>
</ol>
And then stops. Then this becomes one transaction. Now suppose step 3 failed, then our system should:
    <ol>
    <li>Undo step 2</li>
    <li>Undo step 1</li>
</ol>
And hence the final state becomes same as before transaction started.<br>This property ensures data integrity</h3>

 <h2 style=" font-family:'Times New Roman';"><b>Consistency Test:</b></h2>

<h3 style=" font-family:'Times New Roman';">Consistency means that after every committed transaction, the database must satisfy all contraints and rules. So a transaction cannot leave a database in an invalid state.<br>
Some of the common contraints in consistency test are:
<ol>
    <li>Primary Key Uniqueness</li>
    <li>Foreign Key references must exist</li>
    <li>No invalid values (like age <=0)</li>
</ol>
For simplicity, we have implemented majorly the Primary key Uniqueness constraint and the update error, i.e., updating a value that doesn't exist otherwise the table.py would become too specific for tables as different tables have different constraints. Hence here we have implemented primary key uniqueness and will demonstrate that.
<br><br><b>Note:</b> 
Since the system is designed as a generic table abstraction backed by a B+ Tree index, implementing table-specific constraints (such as CPI limits or foreign key relationships between specific tables) would make the table layer schema-dependent and reduce its generality.<br>
Therefore, for this implementation we enforce the primary key uniqueness constraint as well as take care that there are no wrong updates, which is universally applicable across relations. This constraint ensures that duplicate records cannot be inserted into any table.<br></h3>

<h3 style=" font-family:'Times New Roman';"><b>Test - I: Primary Key Uniqueness</b></h3>

#### Initial Database:

```python
print("Initial Database State")

student.show()
course.show()
enrollment.show()
```

```python
# Beginning the transaction
print("\n--- Consistency Test ---")

tid = tm.begin()

try:
    print("Operation 1 → Insert Student")
    student.insert((230020,"23AI010",2023,8.2), tm, tid)

    print("Operation 2 → Update Course")
    course.update(101, {"LTPC":"3_0_0_4"}, tm, tid)

    print("Operation 3 → Insert Enrollment with duplicate key")
    enrollment.insert((1,230020,101,"2023"), tm, tid)   # 1 already exists --> ERROR as primary key being tried to be duplicated

    tm.commit(tid)

except Exception as e:
    print("Consistency violation detected:", e)
    tm.rollback(tid)
```

<h3 style=" font-family:'Times New Roman';">Here we attempt to violate the primary key constraint in the enrollment table. Initially, the table already contains the record:<br><br>
    <center>(1, 230005, 101, '2024')</center><br>

We then try to insert another record with the same primary key:<br>

<center>(1, 230020, 101, '2023')</center><br>

Since our B+ Tree index uses the first column (Enrollment ID) as the primary key, duplicate keys are not allowed. When the system detects that the key 1 already exists, it raises a duplicate primary key error.<br>

Because this operation is executed inside a transaction, the failure causes the entire transaction to abort. To preserve atomicity, the transaction manager performs a rollback, undoing all previous operations within that transaction.<br>

As a result, the database state remains unchanged, and the table contents before and after the failed transaction are identical.<br>

This demonstrates that the system maintains consistency, since the database never enters an invalid state where primary key constraints are violated.<br><br>
<b>Note:</b>
In this simplified implementation, we enforce consistency primarily through primary key uniqueness and validation in the update operation. More complex constraints such as foreign keys or domain constraints were not implemented to keep the database engine generic and lightweight</h3>

 <h2 style=" font-family:'Times New Roman';"><b>Isolation Test:</b></h2>

<h3 style=" font-family:'Times New Roman';">Isolation means that multiple transactions running concurrently should not interfere with each other. This means even though multiple users are performing transaction on same table, it should happen as if oly one transaction is happening in the system. <br>For example: Suppose Transaction T1 and T2 are running together. Still the system should act as:
<ol>
    <li>First T1 runs and then T2 or</li>
    <LI>First T2 runs and then T1</LI>
</ol>
Hence there should not be situations like transaction see's another transaction intermediate state while performing its own operation.
So to show that our mini database passes the Isolation test, we will run two transactions on same table without doing any commit. This creates the environment that two people trying to access same table at same time.</h3>

<h3 style=" font-family:'Times New Roman';"><b>Expectations:</b><br>So we will here demonstrate dirty read prevention example, i.e., consider we have student table.<br>
Two Transaction:
<ol>
    <li>T1 --> Updates Student CPI</li>
    <li>T2 --> Searches for same student's CPI</li>
</ol>
If isolation works properly, then T2 should not see uncommited data from T1 as both are working on same table. Simple example: T1 updates CPI from 8.2 to 9.5. When T2 searches at same time, it should see 8.2 and not 9.5 as thats not committed till now. Once committed if T2 checks then it can see. Hence there are no intermediate reads (dirty reads)<br><br> 
Hence:<br>
- Multiple transactions may run “together,” but the system behaves as if they were executed one after the other.<br>
- No transaction should ever see another’s intermediate state.
</h3>

 <h2 style=" font-family:'Times New Roman';"><b>Durability Test:</b></h2>

<h3 style="font-family:'Times New Roman';">Durability means once a transaction is committed, its changes must never be lost even if the system crashes. Now we will have doubt that once committed then it is changed to database and hence why is it necessary? 
<BR> This is because sometimes the timeline might look like:
<ol>
    <li>T1 begins</li>
    <Li>T1 performs operation</Li>
    <li>T1 commits</li>
</ol>
But before it could successfully upgrade to the B+ Tree, every thing that could went wrong, went wrong and the system crashed!! Now, the transaction shows its committed but the database is unchanged. And such thing is the last thing that any database designer would want that transaction is committed but no changed happened. It would look like scam. And hence to avoid this scam, we have the fourth and the last property of a database engine as <b>Durability.</b></h3>

<h3 style="font-family:'Times New Roman';">So to avoid this scam, we implement WAL (Write Ahead Logging) that stores all the logs, i.e., operations present in the transaction. So durability have to ensure that before changing database, write change to disk log. Hence the sequence becomes:
<ol>
    <li>Write log to file/disk</li>
    <Li>Apply update</Li>
    <li>Commit</li>
</ol>
If crash happens, then we can read log and then redo (perform) committed operations. Hence by this way committed changes cannot disappear. So as soon as we open the system, we can redo and recover all the commited changes.</h3>

#### Initial Database:

```python
print("Initial Database State")

student.show()
course.show()
enrollment.show()
```

```python
# Now begins the final test: Durability test
print("\n--- Durability Test ---")

# Starting transaction
t1 = tm.begin()
student.update(230005, {"CPI": 9.9}, tm, t1)
course.update(201, {"LTPC":"3_2_2_5"}, tm, t1) 
enrollment.insert((8,230020,101,"2023"), tm, t1)   

# Committing transaction
tm.commit(t1)

#_____________________CRASH____________________#
print("SYSTEM CRASH!!!!!!!!")
# SimulatING crash: reinitializing the database to restart the system. 
db_manager = DB_Manager()   # fresh DB manager
# Now firstly lets see content in wal.log
wal_log = WAL()
print("\nWrite Ahead Log Contents")
print("------------------------")
wal_log.show_logs()
print("------------------------")

print("Now we can use this to get back and recover the Committed data!")
wal_log.recover(db_manager)

# Verify recovery
print("After recovery:", student.search(230005))
# Expected: CPI = 9.5 (committed change survives crash)
```

#### Database after Transaction:

```python
print("After transaction:Database State")

student.show()
course.show()
enrollment.show()
```

<h3 style="font-family:'Times New Roman';">We can clearly see here that once transaction was committed, we restarted the system by reintializing the db_manager. This was done intenionally to bring in failure and system crash as in such scenarios the system restarts. Hence we restarted the db manager. Then once restarted we used the logs that was already stored during committing before performing operations and then perfomed operations in them and by this way we recovered the committed transaction into database which couldn't get written due to crash.
<br><br> Hence, by this way our system handles Durability and passes its test.<br><BR>
<b>Note: </b>In this test, the data appeared to be already written into the database because simulating a true system crash within the Python interpreter is not possible. As soon as commit() is called, Python executes the operations immediately and updates the in‑memory B+ Tree. <br>In real database systems, however, committing large volumes of data may take time, and a crash can occur during that window. This is precisely why Write‑Ahead Logging (WAL) is essential: the commit record is logged before operations are applied, ensuring that even if a crash interrupts the physical write, recovery can replay the committed transaction. Thus, while our demo shows durability conceptually, actual systems rely on WAL and checkpoints to guarantee durability under real crash scenarios.

<h3 style="font-family:'Times New Roman';">So we can see that every time we are reseting the log. This we did only to lessen the complexity of our project so that not many logs appear and it becomes difficult to interpret.<br>
In reality, we never reset WAL after every commit, because WAL is essential for durability: it ensures committed transactions survive crashes until the database pages are flushed. Resetting WAL too early would risk losing committed changes. Instead, WAL is truncated only at checkpoints, when all pages have been written to disk. This design guarantees durability while keeping recovery efficient.<br><BR>
Hence, <b>Resetting WAL after every commit breaks durability; resetting only at checkpoints balances safety and efficiency.</b>
